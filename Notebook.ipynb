```python
df['concat'] = df['title'] + ' ' + df['artist'] + ' ' + df['top genre']
```

```python
df['concat'].head()
```

```python
vt = TfidfVectorizer()
vector = vt.fit_transform(df['concat'])
print(vector)
```

```python
similar_score = cosine_similarity(vector)
print(similar_score)
```

```python
user_in = input("Enter song : ")
```

```python
song_list = list(df['title'])
close_song = d.get_close_matches(user_in, song_list)
print(close_song)
print(close_song[0])
```

```python
close_index = df.loc[df['title'] == close_song[0]].index[0]
print(close_index)
```

```python
similar = list(enumerate(similar_score[close_index]))
similar_sort = sorted(similar, key = lambda x : x[1], reverse = True)
```

```python
print(similar_sort)
```

```python
id_list = []

for i in range(0, 20):
    x = similar_sort[i][0]
    id_list.append(x)

print(id_list)
```

```python
print("Song recommendation :")
k = 1
for i in id_list:
    j = i - 1
    song = df[df['Unnamed: 0'] == j]['title'].values[0]
    print(k, song)
    k = k + 1
```
